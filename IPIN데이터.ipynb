```python
sign = []
for i,row in df_subset1.iterrows() :
    ct = 0
    for j in row :
        if j!= 100 :
            ct += 1
    sign.append(ct)
df_subset1['sign_cnt'] = sign

df_subset1.head()
```

```python
col = ['RELATIVEPOSITION', 'sign_cnt']
df_subset2 = df_subset1[col]
df_subset2
```

```python
%%time
reducer = umap.UMAP(random_state=42, n_components=2, min_dist=0.25, n_epochs=500)
reducer.fit(df_subset2)
embedding = reducer.transform(df_subset2)
```

```python
num_features = embedding.shape[1]
som = SOM(m=2, n=1, dim=num_features, lr=0.1, max_iter=3000, random_state=42)
som.fit(embedding, epochs=100)

cluster_label = som.predict(embedding)
result = df_subset2.copy()
result['relativeposition'] = cluster_label
result['emb_1'] = embedding[:,0]
result['emb_2'] = embedding[:,1]

# 2 dimension
plt.figure(figsize=(8,6))
sns.scatterplot(data = result, x = 'emb_1', y='emb_2', hue='relativeposition' , palette='Pastel2')
plt.title('The Plot Of The Clusters(2D)')
plt.show()
```
